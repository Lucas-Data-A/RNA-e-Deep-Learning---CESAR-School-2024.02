1. Número de camadas (Hidden Layers)
Impacto: Aumentar o número de camadas geralmente permite que a rede aprenda padrões mais complexos.

No exemplo:usando 5 camadas ocultas, o que é adequado para capturar a complexidade do padrão espiral.

Cuidado: Muitas camadas podem levar ao overfitting se os dados forem simples ou se não houver regularização adequada.

2. Número de neurônios por camada
Impacto: Mais neurônios aumentam a capacidade da rede de representar funções complexas.

No exemplo: A rede tem de 2 a 5 neurônios por camada. Isso ajuda a capturar as não linearidades da espiral.

Cuidado: Neurônios demais sem regularização ou dados suficientes também levam a overfitting.

3. Função de ativação (Activation)
Impacto: Define como os sinais são transformados entre as camadas.

No exemplo: Está sendo usada ReLU, que é eficiente e comum em problemas com muitas camadas.

Alternativas: Tanh ou sigmoid são boas para problemas mais simples ou com poucas camadas.

4. Taxa de aprendizado (Learning rate)
Impacto: Controla o tamanho dos passos durante a atualização dos pesos.

No exemplo: Está em 0.03, o que parece estar funcionando bem (perda está baixa e estável).

Cuidado: Taxas muito altas podem fazer a rede oscilar ou não convergir; muito baixas tornam o treino lento ou fazem a rede parar em mínimos ruins.

5. Regularização (L2 + Regularization rate)
Impacto: Penaliza pesos grandes para evitar overfitting.

No exemplo: Está sendo usada L2 com taxa 0.003. Isso ajuda a generalizar melhor, especialmente com muitos neurônios/camadas.

Cuidado: Regularização excessiva pode impedir a rede de aprender padrões importantes.
